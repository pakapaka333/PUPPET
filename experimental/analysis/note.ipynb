```python
import matplotlib.pyplot as plt

def make_graph_title(dataset_name: str, model_name: str, metric_key: str) -> str:
    return f"Comparison of {metric_key} [{dataset_name}] ({model_name})"

def boxplot_metrics(result_dfs: dict[str, pd.DataFrame], metric_key: str, graph_title: str) -> None:
    def __flatten_metrics(metrics_list: list[float] | list[list[float]]) -> list[float]:
        if isinstance(metrics_list[0], float):
            return metrics_list
        else:
            return [item for sublist in metrics_list for item in sublist]

    metrics = {
        setting_name: __flatten_metrics(result_dfs[setting_name][metric_key])
        for setting_name in result_dfs.keys()
        if metric_key in result_dfs[setting_name].columns
    }
    if len(metrics) == 0:
        raise ValueError(f"No metrics found for {metric_key}")

    labels = list(metrics.keys())
    data = list(metrics.values())

    fig_width = max(6, 0.7 * len(labels))
    _, ax = plt.subplots(figsize=(fig_width, 5))
    ax.boxplot(data, vert=True, showmeans=True)
    ax.set_xticks(range(1, len(labels) + 1))
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=8)
    ax.set_title(graph_title)
    ax.set_ylabel(metric_key)

    plt.subplots_adjust(bottom=0.25)
    plt.tight_layout()
    plt.show()


def get_average_score(result_dfs: dict[str, pd.DataFrame], metric_key: str) -> float:
    average_over_list = lambda score_list: sum(score_list)/len(score_list)
    average_scores = {
        setting_name: result_dfs[setting_name][metric_key].apply(average_over_list).mean() if isinstance(result_dfs[setting_name][metric_key].iloc[0], list) \
            else result_dfs[setting_name][metric_key].mean()
        for setting_name in result_dfs.keys()
        if metric_key in result_dfs[setting_name].columns
    }
    return average_scores
```

```python
from sklearn.metrics import roc_curve, auc
from dataclasses import dataclass
from typing import List
import numpy as np
import matplotlib.pyplot as plt


@dataclass
class RocSummary:
    fpr_grid: np.ndarray
    median_tpr: np.ndarray
    tpr_low: np.ndarray      # 2.5 percentile
    tpr_high: np.ndarray     # 97.5 percentile
    aurocs: np.ndarray
    auroc_mean: float
    auroc_ci_low: float      # 2.5 percentile
    auroc_ci_high: float     # 97.5 percentile


def get_roc_summary_with_mean_scores(
    detect_likelihoods_machine: List[List[float]],
    detect_likelihoods_human: List[List[float]],
    true_label: str = "machine",
    num_points: int = 101,
) -> RocSummary:
    """各サンプルのスコアの『平均値』を使って、
    ROC 曲線と AUROC を 1 回だけ計算する。

    既存の RocSummary 型に合わせるため、
    - median_tpr: 平均スコアで計算した ROC の TPR
    - tpr_low, tpr_high: 同じ値（CI は無し）
    - aurocs: 要素 1 個だけ（平均スコアで計算した AUROC）
    - auroc_mean, auroc_ci_low, auroc_ci_high: すべて同じ値
    として返す。
    """

    if true_label not in ["machine", "human"]:
        raise ValueError(f"True label must be 'machine' or 'human', but got {true_label}.")

    # 1: positive (true_label), 0: negative
    machine_cls, human_cls = (1, 0) if true_label == "machine" else (0, 1)

    # 各サンプル集合ごとに単純平均を取る
    mean_scores_machine = [float(np.mean(ls)) for ls in detect_likelihoods_machine if ls]
    mean_scores_human = [float(np.mean(ls)) for ls in detect_likelihoods_human if ls]

    y_true = [machine_cls] * len(mean_scores_machine) + [human_cls] * len(mean_scores_human)
    y_score = mean_scores_machine + mean_scores_human

    # 通常の ROC / AUC
    fpr, tpr, _ = roc_curve(y_true, y_score)
    auroc = float(auc(fpr, tpr))

    # 共通の FPR グリッドを作成し、TPR を線形補間
    fpr_grid = np.linspace(0.0, 1.0, num_points)
    interp_tpr = np.interp(fpr_grid, fpr, tpr)

    # CI は持っていないので、全部同じ値で埋める
    median_tpr = interp_tpr
    tpr_low = interp_tpr
    tpr_high = interp_tpr

    aurocs = np.array([auroc], dtype=float)
    auroc_mean = auroc
    auroc_ci_low = auroc
    auroc_ci_high = auroc

    return RocSummary(
        fpr_grid=fpr_grid,
        median_tpr=median_tpr,
        tpr_low=tpr_low,
        tpr_high=tpr_high,
        aurocs=aurocs,
        auroc_mean=auroc_mean,
        auroc_ci_low=auroc_ci_low,
        auroc_ci_high=auroc_ci_high,
    )


def plot_roc_with_ci(summary: RocSummary, title: str | None = None, target_fpr: float = 0.01):
    """median ROC + 95% CI band を描画し、
    特定の FPR（例: 0.1）における TPR を点として追加する。
    """

    fpr_grid = summary.fpr_grid
    median_tpr = summary.median_tpr

    # --- FPR=target_fpr に最も近い点を取得 ---
    idx = np.argmin(np.abs(fpr_grid - target_fpr))
    fpr_point = fpr_grid[idx]
    tpr_point = median_tpr[idx]

    plt.figure()

    # Median ROC
    plt.plot(
        fpr_grid,
        median_tpr,
        label=(
            f"Median ROC\n"
            f"AUC = {summary.auroc_mean:.3f} "
            f"(95% CI [{summary.auroc_ci_low:.3f}, {summary.auroc_ci_high:.3f}])"
        ),
    )
    print(f"AUC: {summary.auroc_mean}, ", end="")

    # 95% CI band
    plt.fill_between(
        fpr_grid,
        summary.tpr_low,
        summary.tpr_high,
        alpha=0.2,
        label="TPR 95% percentile band",
    )

    # --- FPR = 0.1 の点をプロット ---
    plt.scatter(
        [fpr_point],
        [tpr_point],
        color="red",
        s=50,
        zorder=5,
        label=f"TPR at FPR={fpr_point:.2f} → {tpr_point:.3f}",
    )
    print(f"TPR at FPR={fpr_point}: {tpr_point}")

    # Chance line
    plt.plot([0, 1], [0, 1], linestyle="--", label="Chance")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title or "ROC with 95% CI (random sampling)")
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def get_tpr_at_fpr(summary: RocSummary, target_fpr: float = 0.01):
    fpr_grid = summary.fpr_grid
    median_tpr = summary.median_tpr

    # --- FPR=target_fpr に最も近い点を取得 ---
    idx = np.argmin(np.abs(fpr_grid - target_fpr))
    tpr_point = median_tpr[idx]
    return tpr_point
```

```python
from typing import List, Dict, Any
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def plot_auc_curves(
    true_labels_list: List[np.ndarray],
    pred_scores_list: List[np.ndarray],
    curve_names: List[str],
    title: str = "ROC Curves",
    xlabel: str = "False Positive Rate",
    ylabel: str = "True Positive Rate",
) -> List[Dict[str, Any]]:
    # --- 入力チェック ---
    if not (len(true_labels_list) == len(pred_scores_list) == len(curve_names)):
        raise ValueError("true_labels_list, pred_scores_list, curve_names の長さは同じである必要があります。")

    results: List[Dict[str, Any]] = []

    plt.figure(figsize=(6, 6))

    for y_true, y_score, name in zip(true_labels_list, pred_scores_list, curve_names):
        y_true = np.asarray(y_true)
        y_score = np.asarray(y_score)

        if y_true.shape[0] != y_score.shape[0]:
            raise ValueError(f"{name}: y_true と y_score の長さが一致しません。")

        fpr, tpr, thresholds = roc_curve(y_true, y_score)
        roc_auc = auc(fpr, tpr)

        # Youden J = TPR - FPR の最大点
        youden = tpr - fpr
        best_idx = int(np.argmax(youden))
        best_fpr, best_tpr, best_thr = fpr[best_idx], tpr[best_idx], thresholds[best_idx]

        # 描画（色指定なし＝matplotlib のデフォルト配色）
        plt.plot(fpr, tpr, label=f'{name} (AUC={roc_auc:.3f})')
        print(f"AUC: {roc_auc}")

        plt.scatter([best_fpr], [best_tpr], edgecolors='black', zorder=5,
                    label=f'{name} Youden (thr={best_thr:.2f})')

        results.append({
            "name": name,
            "auc": float(roc_auc),
            "best_threshold": float(best_thr),
            "best_fpr": float(best_fpr),
            "best_tpr": float(best_tpr),
            "fpr": fpr,
            "tpr": tpr,
            "thresholds": thresholds,
        })

    # 対角線（ランダム）
    plt.plot([0, 1], [0, 1], 'k--', label='Random')

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return results


from sklearn.metrics import roc_curve, accuracy_score

def accuracy_at_fpr(
    y_true: np.ndarray,
    y_score: np.ndarray,
    target_fpr: float = 0.1
) -> dict:
    fpr, tpr, thresholds = roc_curve(y_true, y_score)

    # target_fpr に一番近い閾値を選ぶ
    idx = int(np.argmin(np.abs(fpr - target_fpr)))
    thr = thresholds[idx]

    # その閾値で予測を作成
    y_pred = (y_score >= thr).astype(int)
    acc = accuracy_score(y_true, y_pred)

    return {
        "target_fpr": target_fpr,
        "chosen_threshold": float(thr),
        "actual_fpr": float(fpr[idx]),
        "tpr": float(tpr[idx]),
        "accuracy": float(acc),
    }
```

```python
z_score_key = "score"

higher_is_machine = {
    "KGW": True,
    "SynthID": True,
    "EXPGumbel": False,
    "Unigram": True,
    "K-SemStamp": True,
}

for setting_name in detection_result_dfs.keys():
    det_df = detection_result_dfs[setting_name]

    try:
        valid_z_score_human = list(filter(lambda x: not np.isnan(x), det_df['Human'][z_score_key].tolist()))
    except:
        print(f"☠️ Error: Human: {setting_name} has no z-score")
        continue

    try:
        valid_z_score_machine = list(filter(lambda x: not np.isnan(x), det_df['Watermarked'][z_score_key].tolist()))
    except:
        print(f"☠️ Error: Watermarked: {setting_name} has no z-score")
        continue

    if len(valid_z_score_machine) != len(valid_z_score_human):
        print(f"⚠️ Warning: {setting_name} has different length of valid z-score - Human: {len(valid_z_score_human)}, Watermarked: {len(valid_z_score_machine)}")

    pred_score_mgpositive = np.concatenate((
        valid_z_score_machine, 
        valid_z_score_human
        ))

    setting_name_core = setting_name.split()[0].strip()
    if higher_is_machine[setting_name_core]:
        # 1: machine-generated, 0: human-written
        true_label_mgpositive = np.concatenate((
            np.ones_like(valid_z_score_machine, dtype=int), 
            np.zeros_like(valid_z_score_human,  dtype=int)
            ))
    else:
        # 0: machine-generated, 1: human-written
        true_label_mgpositive = np.concatenate((
            np.zeros_like(valid_z_score_machine, dtype=int), 
            np.ones_like(valid_z_score_human,  dtype=int)
            ))

    results = plot_auc_curves(
        true_labels_list=[true_label_mgpositive],
        pred_scores_list=[pred_score_mgpositive],
        curve_names=['Watermarked'],
        title=f"ROC Curves [{task_name}] ({model_name}. {setting_name})",
    )

    result_mg = accuracy_at_fpr(true_label_mgpositive, pred_score_mgpositive, target_fpr=0.01)
    print(f"{setting_name}: TPR@FPR=0.01: {result_mg['tpr']}")
```
